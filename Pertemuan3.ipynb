```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

## Read data

```python
df = pd.read_csv("House_Rent_Dataset.csv")
df.head(10)
```

```python
# Matplotlib
x=list(df["Area Type"])
y=list(df["Size"])

plt.bar(x,y,color='red')
plt.show()
```

```python
# Seaborn
plt.figure(figsize=(10,5))
sns.countplot(data=df,x="City",palette='flare',hue="Area Type")
plt.show()
```

```python
# Matplotlib
x=list(df["Area Type"])
y=list(df["Size"])

plt.barh(x,y,color='red')
plt.show()
```

```python
# Grafik Garis
x=list(df["Posted On"])
y=list(df["Rent"])

plt.plot(x,y)
plt.show()
```

```python
df["Posted On"]=pd.to_datetime(df["Posted On"])
df.dtypes
```

```python


monthly_rent=df.resample('M',on='Posted On').mean()
monthly_rent.head()
```

```python

plt.figure(figsize=(10,10))
sns.lineplot(data=monthly_rent,x=monthly_rent.index,y='Rent')
plt.title("Kenaikan Harga Sewa Bulanan")
plt.show()
```

# Grafik Pie

```python
data_pie=df["Area Type"].value_counts()
data_pie.head()
```

```python
data_pie=df["Area Type"].value_counts().rename_axis("Area Type")
data_pie.head()
```

```python
data_pie=df["Area Type"].value_counts().rename_axis("Area Type").reset_index(name="Id_Count")
data_pie.head()
```

```python
plt.pie(
    data_pie["Id_Count"],
    labels=data_pie["Area Type"],
    startangle=90,autopct='%.2f%%',
    explode=(0.1,0,0),shadow=True)

plt.show()
```

# Scatter Plot
untuk membandingkan  dua atribut numerik

```python
#Matplotlib
plt.figure(figsize=(10,5))
plt.scatter(x=df["Rent"],y=df["Size"])
plt.show()
```

```python
#Seaborn
sns.set_style("darkgrid")
plt.figure(figsize=(10,8))
sns.scatterplot(data=df,x='',y='Size',hue='City')
plt.show()
```

# BoxPlot
untuk melihat penyebaran data dengan atribut kategorikal terhadap data dengan atribut numerik (Category Vs Numerik)

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df,x="Size",y="City",orient='h')
plt.show()
```

# Heatmap

```python
plt.figure(figsize=(10,5))
sns.heatmap(
    df.corr(),
    annot=True,
    cmap="flare"
)
plt.show()
```
